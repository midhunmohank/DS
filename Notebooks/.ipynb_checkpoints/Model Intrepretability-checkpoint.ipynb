### 1. Fit a linear model and interpret the regression coefficients

### Linear Model

```python
import statsmodels.api as sm

lin_reg = sm.OLS(y_train,X_train).fit()

```

```python
lin_reg.summary()
```

```python
#Predicting the values
lr_pred = lin_reg.predict(X_test)
```

```python
#Checking accuracy of predictions
print('Linear Model - Test')
print('MAE: %f'% mean_absolute_error(y_test, lr_pred))
print('RMSE: %f'% np.sqrt(mean_squared_error(y_test, lr_pred)))   
print('R2 %f' % r2_score(y_test, lr_pred))
```

#### Interpretation of model coefficients:

1. Prob(F- Statistic) = 0.00 indicates that our model is satatistically significant an all significace levels.
2. R squared value of .317 indicate that 31.7% of the data is explained/fit by the model.
3. P > |t| has a value less than 0.1 for logitude, latitude, neighbourhood group, neighbourhood, room type, no of reviews, availabilty 365 indicating that these variables are important in predicting the price.


### 2. Fit a tree-based model and interpret the nodes

### Tree based Model

```python
from sklearn.tree import DecisionTreeRegressor
DTree=DecisionTreeRegressor(max_depth=3, random_state=1234)
DTree.fit(X_train,y_train)

DTree_score = DTree.score(X_test, y_test)
print(DTree_score)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error
def model_sel(model):
    y_pred = model.predict(X_test)
    print("MODEL EVALUATION")
    print("\nr2_score of the model: ")
    print(r2_score(y_test,y_pred))
```

```python
# Model Evaluation on DecisionTreeRegressor
print("DECISION TREE REGRESSION: ")
print("-x"*25)
model_sel(DTree)
```

```python
prediction_tree = DTree.predict(X_test)
```

```python
#Checking accuracy of predictions
print('Decision tree')
print('MAE: %f'% mean_absolute_error(y_test, prediction_tree))
print('RMSE: %f'% np.sqrt(mean_squared_error(y_test, prediction_tree)))   
print('R2 %f' % r2_score(y_test, prediction_tree))
```

```python
from sklearn import tree
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(25,20))
tree.plot_tree(DTree, feature_names= features.columns, filled=True)
```

The max_depth of the tree is set at 3. With each note, the data is being split into smaller subset. Node 0 has all samples of the dataset, but it'll be lower by the Node 3.
Node 0 - room_type - took 32759 samples to arrive at a price value of 154.313$
Node 2 and Node 3 - longitude, minumum nights - took 17039 and 32514 samples to set a bracket for availability, longitide and neighbourhood and further the leaves of the tree has a bracket for mse, wherein each mse bracket is given a price value.

### 4. Run SHAP analysis on the models from steps 1, 2, and 3,   interpret the SHAP values and compare them with the other model interpretability methods.

## Shap Analysis

SHAP (SHapley Additive exPlanations) by Lundberg and Lee (2017)69 is a method to explain individual predictions. SHAP is based on the game theoretically optimal Shapley values.
The goal of SHAP is to explain the prediction of an instance x by computing the contribution of each feature to the prediction. The SHAP explanation method computes Shapley values from coalitional game theory.

```python
!pip3 install shap
```

```python
import shap
```

### 1. Shap Analysis of the Linear Model

```python
explainer_sh = shap.LinearExplainer(lin_reg, X_train, feature_dependence='independent')
shap_values_lg = explainer_sh.shap_values(X_test)
shap.summary_plot(shap_values_lg, X_test, feature_names=features.columns)
```

```python
shap.summary_plot(shap_values_lg, X_train, plot_type="bar", color='steelblue')
```

The room type is the most significant feature followed by the availability of the Airbnb in a specific logitude at a famous neighbourhood.

### 2. Tree Model - Shap Analysis

```python
explainer_tree = shap.Explainer(DTree)
shap_values_tree = explainer_tree(X_test)
```

```python
shap.summary_plot(shap_values_tree, X_test, feature_names=features.columns)
```

The tree model suggests that room type and location is more signifiacnt where in in linear regression, room and availability were the most significant features.
